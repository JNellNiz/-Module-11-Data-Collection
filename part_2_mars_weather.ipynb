```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
page_soup = soup(html, 'html.parser')
```

```python
# Extract all rows of data

# Find the table that contains the data
table = page_soup.find('table')

# Extract the column names from the table header
column_names = [th.text for th in table.find_all('th')]
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
data_rows = []

# Loop through the scraped data to create a list of rows
for tr in table.find_all('tr')[1:]:
    data_row = [td.text for td in tr.find_all('td')]
    data_rows.append(data_row)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data_rows, columns=column_names)
```

```python
# Confirm DataFrame was created successfully
print(df)
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
print(df.dtypes)
```

```python
# Change data types for data analysis
# Convert 'terrestrial_date' column to datetime
df['terrestrial_date'] = pd.to_datetime(df['terrestrial_date'])

# Convert 'sol' column to integer
df['sol'] = df['sol'].astype(int)

# Convert 'ls' column to integer
df['ls'] = df['ls'].astype(int)

# Convert 'month' column to integer
df['month'] = df['month'].astype(int)

# Convert 'min_temp' column to float
df['min_temp'] = df['min_temp'].astype(float)

# Convert 'pressure' column to float
df['pressure'] = df['pressure'].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
print(df.dtypes)
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
month_counts = df['month'].value_counts()

# Print the result
print(month_counts)
```

```python
# 2. How many Martian days' worth of data are there?
# Count the number of unique Martian days
num_martian_days = df['sol'].nunique()

# Print the result
print("The number of Martian days' worth of data is:", num_martian_days)
```

```python
# 3. What is the average low temperature by month?
# Calculate the average low temperature by month
average_low_temp_by_month = df.groupby('month')['min_temp'].mean()

# Print the result
print("Average low temperature by month:")
print(average_low_temp_by_month)
```

```python
# Plot the average temperature by month

# Calculate the average low temperature by month
average_temp_by_month = df.groupby('month')['min_temp'].mean()

# Plot the average temperature by month
plt.bar(average_temp_by_month.index, average_temp_by_month)
plt.xlabel('Month')
plt.ylabel('Average Low Temperature (Celsius)')
plt.title('Average Low Temperature on Mars by Month')
plt.show()
```

```python
# Sort the average temperature values in ascending order
average_temp_sorted = average_temp_by_month.sort_values()

# Create a list of sorted month labels based on the sorted temperature values
month_labels_sorted = [str(month) for month in average_temp_sorted.index]

# Plot the average temperature by month with sorted bars
plt.bar(month_labels_sorted, average_temp_sorted)
plt.xlabel('Month')
plt.ylabel('Average Low Temperature (Celsius)')
plt.title('Average Low Temperature on Mars by Month (Sorted by Temperature)')
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location

# Calculate the average low temperature by month
average_low_temp_by_month = df.groupby('month')['min_temp'].mean()

# Sort the average low temperatures in ascending order
sorted_temps = average_low_temp_by_month.sort_values()

# Identify the coldest and hottest months
coldest_months = sorted_temps.head(1)
hottest_months = sorted_temps.tail(1)

# Plot the coldest and hottest months on a bar chart
plt.figure(figsize=(8, 6))
plt.bar(average_low_temp_by_month.index, average_low_temp_by_month.values, color='lightblue', label='Average Low Temperature')
plt.bar(coldest_months.index, coldest_months.values, color='blue', label='Coldest Month')
plt.bar(hottest_months.index, hottest_months.values, color='red', label='Hottest Month')
plt.xlabel('Month')
plt.ylabel('Average Low Temperature')
plt.title('Coldest and Hottest Months in Curiosity\'s Location')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# 4. Average pressure by Martian month

# Calculate the average atmospheric pressure by month
average_pressure_by_month = df.groupby('month')['pressure'].mean()

print(average_pressure_by_month)
```

```python
# Plot the average pressure by month
plt.bar(average_pressure_by_month.index, average_pressure_by_month)
plt.xlabel('Month')
plt.ylabel('Average Atmospheric Pressure (hPa)')
plt.title('Average Atmospheric Pressure on Mars by Month')
plt.show()
```

```python
# Sort the average pressure values in ascending order
average_pressure_sorted = average_pressure_by_month.sort_values()

# Create a list of sorted month labels based on the sorted pressure values
month_labels_sorted = [str(month) for month in average_pressure_sorted.index]

# Plot the average pressure by month with sorted bars
plt.bar(month_labels_sorted, average_pressure_sorted)
plt.xlabel('Month')
plt.ylabel('Average Atmospheric Pressure (hPa)')
plt.title('Average Atmospheric Pressure on Mars by Month (Sorted by Pressure)')
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?

# Calculate the number of terrestrial days since the first data point
df['terrestrial_days'] = (df['terrestrial_date'] - df['terrestrial_date'].min()).dt.days

# Plot the number of terrestrial days vs. the minimum temperature
plt.plot(df['terrestrial_days'], df['min_temp'])
plt.xlabel('Number of Terrestrial Days')
plt.ylabel('Minimum Temperature (Celsius)')
plt.title('Daily Minimum Temperature on Mars')
plt.xticks(rotation=45)
plt.show()

# Calculate the estimated number of terrestrial days in a Martian year
earth_days_in_martian_year = df['terrestrial_days'].max()
print(f"Estimated number of terrestrial days in a Martian year: {earth_days_in_martian_year}")
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
